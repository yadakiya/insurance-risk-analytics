```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("../data/insurance_data.csv")
df.head()
```

```python
df["Margin"] = df["TotalPremium"] - df["TotalClaims"]
df["HasClaim"] = df["TotalClaims"] > 0
```

```python
def claim_frequency(data):
    return data["HasClaim"].mean()

def claim_severity(data):
    return data[data["TotalClaims"] > 0]["TotalClaims"].mean()

def margin(data):
    return data["Margin"].mean()
```

```python
def t_test(group1, group2, column):
    return stats.ttest_ind(group1[column], group2[column], equal_var=False)
```

```python
gauteng = df[df["Province"] == "Gauteng"]
western = df[df["Province"] == "Western Cape"]
```

```python
t_stat, p_val = t_test(gauteng, western, "TotalClaims")
print("Province p-value:", p_val)
```

```python
if p_val < 0.05:
    print("Reject H0 → Provinces are different risk")
else:
    print("Fail to reject H0 → No strong difference")
```

```python
zip1 = df[df["ZipCode"] == df["ZipCode"].value_counts().index[0]]
zip2 = df[df["ZipCode"] == df["ZipCode"].value_counts().index[1]]
```

```python
t_stat, p_val = t_test(zip1, zip2, "TotalClaims")
print("ZipCode p-value:", p_val)
```

```python
if p_val < 0.05:
    print("Reject H0 → ZipCodes differ")
else:
    print("Fail to reject H0")
```

```python
male = df[df["Gender"] == "Male"]
female = df[df["Gender"] == "Female"]
```

```python
t_stat, p_val = t_test(male, female, "TotalClaims")
print("Gender p-value:", p_val)
```

```python
t_stat, p_val = t_test(male, female, "TotalClaims")
print("Gender p-value:", p_val)
```
